# Other examples

```python
import matplotlib.pyplot as plt
import networkx as nx
import numpy as np
from PIL import Image
from scipy.misc import imsave
from src import helpFunctions
```

```python
imBrain1 = Image.open('./images/raw/brain_transverse.jpg')
imBrain2 = Image.open('./images/raw/brain_sagittal.jpg')
imLions  = Image.open('./images/raw/lions.jpg')
imOwl    = Image.open('./images/raw/owl.jpg')

length, width = 100, 100

imBrain1 = imBrain1.resize((width, length), Image.ANTIALIAS)
imBrain2 = imBrain2.resize((width, length), Image.ANTIALIAS)
imLions  = imLions.resize((width, length), Image.ANTIALIAS)
imOwl    = imOwl.resize((width, length), Image.ANTIALIAS)

imBrain1 = np.asarray(imBrain1)
imBrain2 = np.asarray(imBrain2)
imLions  = np.asarray(imLions)
imOwl    = np.asarray(imOwl)

plt.figure(figsize=(10,10))
plt.subplot(2, 2, 1)
plt.imshow(imBrain1)
plt.title('Brain image transverse cut')

plt.subplot(2, 2, 2)
plt.imshow(imBrain2)
plt.title('Brain image sagittal cut')

plt.subplot(2, 2, 3)
plt.imshow(imLions)
plt.title('Lions')

plt.subplot(2, 2, 4)
plt.imshow(imOwl)
plt.title('Owl')

plt.show()
```

## Visualizing intensity histogram

```python
plt.figure(figsize=(10,10))
plt.subplot(2, 2, 1)
helpFunctions.drawHistogram(imBrain1, 'Brain image transverse cut')

plt.subplot(2, 2, 2)
helpFunctions.drawHistogram(imBrain2, 'Brain image sagittal cut')

plt.subplot(2, 2, 3)
helpFunctions.drawHistogram(imLions, 'Lions')

plt.subplot(2, 2, 4)
helpFunctions.drawHistogram(imOwl, 'Owl')

plt.show()
```
